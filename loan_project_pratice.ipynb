```python
#Data Cleaning
```

```python
# summarising number of missing values in each column
loan.isnull().sum()
```

```python
# percentage of missing values in each column
round(loan.isnull().sum()/len(loan.index), 2)*100
```

```python
# removing the columns having more than 90% missing values
missing_columns = loan.columns[100*(loan.isnull().sum()/len(loan.index)) > 90]
print(missing_columns)
```

```python
loan = loan.drop(missing_columns, axis=1)
print(loan.shape)
```

```python
# summarise number of missing values again
100*(loan.isnull().sum()/len(loan.index))
```

```python
# There are now 2 columns having approx 32 and 64% missing values - 
# description and months since last delinquent

# let's have a look at a few entries in the columns
loan.loc[:, ['desc', 'mths_since_last_delinq']].head()
```

```python
# dropping the two columns
loan = loan.drop(['desc', 'mths_since_last_delinq'], axis=1)
```

```python
# summarise number of missing values again
100*(loan.isnull().sum()/len(loan.index))
```

```python
# missing values in rows
loan.isnull().sum(axis=1)
```

```python
# checking whether some rows have more than 5 missing values
len(loan[loan.isnull().sum(axis=1) > 5].index)
```

```python
loan.info()
```

```python
# The column int_rate is character type, let's convert it to float
loan['int_rate'] = loan['int_rate'].apply(lambda x: pd.to_numeric(x.split("%")[0]))
```

```python
# checking the data types
loan.info()
```

```python
loan['emp_length']
```

```python
# also, lets extract the numeric part from the variable employment length

# first, let's drop the missing values from the column (otherwise the regex code below throws error)
loan = loan[~loan['emp_length'].isnull()]

# using regular expression to extract numeric values from the string
import re
loan['emp_length'] = loan['emp_length'].apply(lambda x: re.findall('\d+', str(x))[0])

# convert to numeric
loan['emp_length'] = loan['emp_length'].apply(lambda x: pd.to_numeric(x))
```

```python
loan['emp_length'].value_counts()
```

```python
# looking at type of the columns again
loan.info()
```

```python
#Data Analysis
```

```python
#UNIVARIATE ANALYSIS
```

```python
# Loan Amount Distribution
plt.figure(figsize=(8,5))
sns.histplot(loan["loan_amnt"], bins=40, kde=True, color="skyblue")
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Annual Income Distribution
plt.figure(figsize=(8,5))
sns.histplot(loan["annual_inc"], bins=50, kde=True, color="green")
plt.xlim(0, loan["annual_inc"].quantile(0.95))  # cap at 95th percentile
plt.title("Annual Income Distribution (Capped at 95th Percentile)")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Loan Term Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="term", data=loan, palette="muted")
plt.title("Loan Term Distribution")
plt.xlabel("Term")
plt.ylabel("Count")
plt.show()
```

```python
# Loan Status Distribution
plt.figure(figsize=(8,5))
sns.countplot(y="loan_status", data=loan, order=loan["loan_status"].value_counts().index, palette="Set2")
plt.title("Loan Status Counts")
plt.ylabel("Loan Status")
plt.xlabel("Count")
plt.show()
```

```python
# Purpose of Loan
plt.figure(figsize=(10,6))
loan["purpose"].value_counts().head(10).plot(kind="barh", color="orange")
plt.title("Top 10 Loan Purposes")
plt.xlabel("Count")
plt.ylabel("Purpose")
plt.show()
```

```python
# Interest Rate Distribution
plt.figure(figsize=(8,5))
sns.histplot(loan["int_rate"], bins=30, kde=True, color="purple")
plt.title("Interest Rate Distribution")
plt.xlabel("Interest Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
#BIVARIATE ANALYSIS
```

```python
# Loan Amount by Term
plt.figure(figsize=(7,5))
sns.boxplot(x="term", y="loan_amnt", data=loan, palette="coolwarm")
plt.title("Loan Amount by Term")
plt.show()
```

```python
# Loan Amount vs Annual Income
plt.figure(figsize=(9,6))
sns.scatterplot(x="annual_inc", y="loan_amnt", hue="loan_status", data=loan, alpha=0.5)
plt.xlim(0, loan["annual_inc"].quantile(0.95))
plt.title("Annual Income vs Loan Amount (Capped at 95th Percentile)")
plt.xlabel("Annual Income")
plt.ylabel("Loan Amount")
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```

```python
# Interest Rate vs Loan Amount
plt.figure(figsize=(8,5))
sns.scatterplot(x="int_rate", y="loan_amnt", data=loan, alpha=0.6, color="purple")
plt.title("Interest Rate vs Loan Amount")
plt.show()
```

```python
# Interest Rate by Grade
if "grade" in loan.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x="grade", y="int_rate", data=loan, palette="Set1")
    plt.title("Interest Rate by Grade")
    plt.show()
```

```python
# Annual Income by Loan Status
plt.figure(figsize=(8,5))
sns.boxplot(x="loan_status", y="annual_inc", data=loan, palette="cool")
plt.ylim(0, loan["annual_inc"].quantile(0.95))  # cap for better visualization
plt.title("Annual Income by Loan Status (Capped at 95th Percentile)")
plt.show()
```

```python
# Loan Amount vs Purpose
plt.figure(figsize=(10,6))
sns.boxplot(y="purpose", x="loan_amnt", data=loan, palette="Set3")
plt.title("Loan Amount by Purpose")
plt.show()
```

```python
#CORRELATION ANALYSIS
```

```python
plt.figure(figsize=(12,7))
corr = loan.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
# Top correlated features with loan amount
print("Top correlations with Loan Amount:")
print(corr["loan_amnt"].sort_values(ascending=False).head(10))
```
